```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Definitions
data_file = './data/histogram_rows.csv'
```

```python
#Read data from CSV file
data = pd.read_csv(data_file, sep=',', header=None)
```

```python
x_data = data[0]
y_data = data[1]
```

```python
#Histogram for PGM image

plt.figure(figsize=(15, 5))

plt.hist(x_data, bins=256, range=(0, 256), weights=y_data)

plt.show()
```
